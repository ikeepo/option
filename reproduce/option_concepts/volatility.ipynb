### Volitality
`σ` is the output of BS, one standard deviation annual which put context around iv, high iv allows us to move very far from the stock price while still covering a 1SD range;    
When we look at the iv we're really looking at the expressed one standard deviation percentage on an annual basis so a quick example, if I've got a $50 stock with 20% iv, it means over the course of a year, it would range -10 to 10 with 68% probability.    

`σ` reflects the expected annualized volatility of the underlying asset's logarithmic return (rather price itself) over the option's life, as implied by the option's market price. It assumes returns follow a lognormal distribution. 

 

## `σ` translated to price movement standard deviation
The standard deviation of the asset price in the BS framework(lognormal model) is approximately:  
$$
std\space of \space Price \approx S * \sigma * \sqrt{T}
$$

## How to understand Standard Deviation here
When we are looking at two standard deviations we're pretty much capturing the essence of what is implied or what could potentially happen which is why we look at these values in the option market.   
We like to look at one standard deviation ranges when we're playing new trades,   
and usually when we look at two deviation levels it's usually an indication of perceived max loss, like when a brokerage is looking at calculating a buying power reduction or a margin on a certain position, they might calculate it from two stds and gauge it from there depending on a person's profile. 

### Volatility Surface
> [1](https://www.investopedia.com/articles/stock-analysis/081916/volatility-surface-explained.asp)    

The example refers to the implied volatility of the STOCK, rather the OPTION.  
Every option strike has its own implied volatility. Different IVs comes from :  
- Moneyness: OTM or ITM options have higher or lower IVs compared to ATM which is called "volatility smile".  
- Time to Expiration   
- Market Dynamics: supply and demand for specific options can influence their IV.   
The VS is a tool to visualize how implied volatility - the market's expectation of an asset's price volatility - changes across different option strike prices and expiration dates.  
> It's the source of many [option trading strategies](/strategy/option/implied_volatility_arbitrage.ipynb).
